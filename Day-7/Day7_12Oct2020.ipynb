#### Decision tree Regressor
- gives better results for non linear data

```python
import pandas as pd
df=pd.read_csv("1000_companies.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
x1=df["R&D Spend"]
y1=df["Profit"]
plt.scatter(x1,y1)
plt.show()
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["State"]=le.fit_transform(df["State"])
x=df.iloc[:,0:4]
y=df["Profit"]

from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x,y)
```

```python
y_predict=model.predict(x)
y_predict
```

```python
from sklearn.metrics import r2_score
print(r2_score(y,y_predict))
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure()
tree.plot_tree(model,feature_names=x.columns)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=25)
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
r2_score(y_test,y_predict)
```

#### Random Forest
- in decision tree -overfitting----model get trained well
- randomforest will overcome that overfitting in decision tree
- gives better accuracy when compared to other algorithms
- it works better if there is also missing values

***Random Forest Classifier***

```python
df=pd.read_csv("bird.csv")
df.head()
```

```python
df.isna().sum()
```

```python
import numpy as np
```

```python
df=df.replace(np.nan,0.0)
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["type"]=le.fit_transform(df["type"])
df.head()
```

```python
df["type"].unique()
```

```python
x=df.iloc[:,1:11]
y=df["type"]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=25)
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

***Random Forest Regressor***
